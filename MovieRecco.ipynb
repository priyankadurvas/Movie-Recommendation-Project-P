```python
#converting a column of JSON strings to a column of strings
import json

def column_to_string(df, column):
    df[column] = df[column].apply(json.loads)
    for index, i in zip(df.index, df[column]):
        list1 = []
        for j in range(len(i)):
            list1.append((i[j]['name'])) # the key 'name' contains the name of the genre
        df.loc[index, column] = str(list1)
    return df

```

```python
def get_top_genres(df, column_name):
    # Clean and extract genres
    for i,j in zip(df[column_name], df.index):
        list2=[]
        list2=i
        list2.sort()
        df.loc[j,column_name]=str(list2)
    df[column_name] = df[column_name].str.strip('[]').str.replace(' ','').str.replace("'",'')
    df[column_name] = df[column_name].str.split(',')

    # Get list of unique genres
    genreList = []
    for index, row in df.iterrows():
        genres = row[column_name]
        for genre in genres:
            if genre not in genreList:
                genreList.append(genre)

    # Get top 10 genres
    top_genres = pd.Series(genreList).value_counts()[:10].sort_values(ascending=True)
    
    return top_genres.index.tolist()
top_genres = get_top_genres(movies, 'genres')
print(top_genres)
```

```python
def clean_keywords(dataframe):
    dataframe['keywords'] = dataframe['keywords'].str.strip('[]').str.replace(' ','').str.replace("'",'').str.replace('"','')
    dataframe['keywords'] = dataframe['keywords'].str.split(',')
    
    for i,j in zip(dataframe['keywords'],dataframe.index):
        list2 = []
        list2 = i
        dataframe.loc[j,'keywords'] = str(list2)
    
    dataframe['keywords'] = dataframe['keywords'].str.strip('[]').str.replace(' ','').str.replace("'",'')
    dataframe['keywords'] = dataframe['keywords'].str.split(',')
    
    for i,j in zip(dataframe['keywords'],dataframe.index):
        list2 = []
        list2 = i
        list2.sort()
        dataframe.loc[j,'keywords'] = str(list2)
    
    dataframe['keywords'] = dataframe['keywords'].str.strip('[]').str.replace(' ','').str.replace("'",'')
    dataframe['keywords'] = dataframe['keywords'].str.split(',')
    
    return dataframe

movies = clean_keywords(movies)
```

```python
#calculating the cosine distance between their genre, cast, director, and keyword for two movie ID's
from scipy import spatial
import pandas as pd
#movies = pd.read_csv('movies.csv')

def Similarity(movieId1, movieId2):
    a = movies.iloc[movieId1]
    b = movies.iloc[movieId2]
    
    genresA = a['genres_bin']
    genresB = b['genres_bin']
    
    genreDistance = spatial.distance.cosine(genresA, genresB)
    
    scoreA = a['cast_bin']
    scoreB = b['cast_bin']
    scoreDistance = spatial.distance.cosine(scoreA, scoreB)
    
    directA = a['director_bin']
    directB = b['director_bin']
    directDistance = spatial.distance.cosine(directA, directB)
    
    wordsA = a['words_bin']
    wordsB = b['words_bin']
    wordsDistance = spatial.distance.cosine(directA, directB)
    return genreDistance + directDistance + scoreDistance + wordsDistance

Similarity(5,130)   
#similarity between two vectors, the smaller the cosine distance, the more similar the vectors
```

```python
# putting it all together for the binary labelling
# This wtakes a long time to run at once
def process_data(movies):
    # Process genres
    genreList = []
    for index, row in movies.iterrows():
        genres = row["genres"]
        genres = [genre.strip() for genre in genres]
        for genre in genres:
            if genre not in genreList:
                genreList.append(genre)
        movies.at[index, "genres"] = genres
    for genre in genreList:
        movies[genre] = movies.apply(lambda row: 1 if genre in row["genres"] else 0, axis=1)
    
    # Process cast
    castList = []
    for index, row in movies.iterrows():
        cast = row["cast"]
        cast = [member.strip() for member in cast]
        for member in cast:
            if member not in castList:
                castList.append(member)
        movies.at[index, "cast"] = cast
    for member in castList:
        movies[member] = movies.apply(lambda row: 1 if member in row["cast"] else 0, axis=1)
    
    # Process director
    directorList = []
    for index, row in movies.iterrows():
        director = row["director"]
        if director not in directorList:
            directorList.append(director)
    for director in directorList:
        movies[director] = movies.apply(lambda row: 1 if director == row["director"] else 0, axis=1)
    
    # Process keywords
    keywordList = []
    for index, row in movies.iterrows():
        keywords = row["keywords"]
        keywords = [keyword.strip() for keyword in keywords]
        for keyword in keywords:
            if keyword not in keywordList:
                keywordList.append(keyword)
        movies.at[index, "keywords"] = keywords
    for keyword in keywordList:
        movies[keyword] = movies.apply(lambda row: 1 if keyword in row["keywords"] else 0, axis=1)
    
    # Create new_id
    new_id = list(range(0, movies.shape[0]))
    movies["new_id"] = new_id
    
    # Select relevant columns
    movies = movies[['original_title', 'vote_average', 'new_id'] + genreList + castList + directorList + keywordList]
    
    return movies

#movies = process_data(movies)
```

```python
#fetching the movies DataFrame for the closest matching movie, and then returning a list of the top 10 movies that are most similar to the selected movie based on a combination of genre, cast, director, and keywords

import operator

def predict_score(name):
    new_movie = movies.loc[movies['original_title'].str.contains(name)].iloc[0]
    print('Selected Movie: ', new_movie['original_title'])
    
    def get_neighbors(base_movie, k):
        distances = [(movie['new_id'], Similarity(base_movie['new_id'], movie['new_id'])) for _, movie in movies.iterrows() if movie['new_id'] != base_movie['new_id']]
        neighbors = sorted(distances, key=operator.itemgetter(1))[:k]
        return neighbors
    
    k = 10
    avg_rating = 0
    neighbors = get_neighbors(new_movie, k)
    
    print('\nRecommended Movies: \n')
    for neighbor in neighbors:
        movie = movies.iloc[neighbor[0]]
        avg_rating += movie['vote_average']
        print(f"{movie['original_title']} | Genres: {str(movie['genres']).strip('[]').replace(' ', '')} | Rating: {movie['vote_average']}")
    
    print('\n')
    avg_rating /= k
    print(f"The predicted rating for {new_movie['original_title']} is: {avg_rating:.2f}")
    print(f"The actual rating for {new_movie['original_title']} is {new_movie['vote_average']:.2f}")
    
#calculate similarity between movies based on their genres, cast, director, and keywords using cosine similarity
#recommends movies based on the k-nearest neighbors to the selected movie 
#and calculates the predicted rating based on the average rating of those neighbors
```
